```python
import matplotlib.pyplot as plt
import numpy
import sepPlot
%matplotlib inline  

twoD=SepVector.getSepVector(Hypercube.hypercube(ns=[10,2]))

twoM=twoD.getNdArray()


for i in range (10):
    twoM[0,i]=i
    twoM[1,i]=20-i
sepPlot.Graph(plt,twoD,label1="time",label2="Amplitude",title="WOW").output()
```
